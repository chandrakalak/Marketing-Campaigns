```python
def outlier_count(data, threshold=100000):
    count = 0
    for i in data:
        if i > threshold:
            count += 1
    print(f"Total number of outliers: {count}")
        
        
```

```python
outlier_count(df['Income'])
```

```python
# USA country purchases vs Rest
df['Is_US']=df['Country']=='US'
sns.boxplot(x='Is_US',y='Total_Purchases',data=df)
plt.title=("Total Purchases:USA vs Others")
plt.show()
           
           
```

* Is_USA = False â†’ Customers from the USA
* Is_USA = True â†’ Customers from all other countries
* Observation:No obvious difference in difference in total purchase behavior between the USA and the rest of the world.

```python
## Let us evaluate the above findings with t test
from scipy.stats import ttest_ind

usa=df[df['Country']=='US']['Total_Purchases']
non_usa=df[df['Country']!='US']['Total_Purchases']

t_stat,p_val=ttest_ind(usa,non_usa,equal_var=False)

print(f"T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}")
```

* no difference in the purchase between US and other countries
